```python
import nibabel as nib
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
# Read all the files in the directory
# and store them in a list
files = glob.glob('data/train/volume/*.nii')
```

```python
liste_caracteristiques = []
for i in range(len(files)):
    img = nib.load(files[i])
    liste_caracteristiques.append([img.shape, img.header.get_zooms(), img.header.get_xyzt_units(), img.header.get_data_dtype()])
```

```python
dataframe = pd.DataFrame(liste_caracteristiques, columns=['shape', 'zooms', 'units', 'dtype'])
```

```python
dataframe.head(30)
```

```python
imag_info = []
for i in range(len(files)):
    img = nib.load(files[i]).get_fdata()
    imag_info.append([np.min(img), np.max(img)])
```

```python
df_img_info = pd.DataFrame(imag_info, columns=['min', 'max'])
```
